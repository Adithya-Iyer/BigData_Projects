```
import nltk
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords
import re
from string import punctuation
```

```
n = summaries.count()
```

```
n
```

```
id_summary = summaries.map(lambda x: x.split('\t'))
```

```
id_summary.collect()
```

```
id_summary_l = id_summary.map(lambda x: [x[0], x[1].lower()])
id_summary_l.take(5)
```

```
id_summaryT = id_summary_l.map(lambda x: [x[0], wordpunct_tokenize(x[1])])
id_summaryT.take(5)
```

```
nltk.download('stopwords')
sw = stopwords.words('english')

def preprocess(tokens):
    tokens = [token for token in tokens if token not in punctuation]
    tokens = [token for token in tokens if token not in sw]
    return tokens

id_summary_lTnoSW = id_summaryT.map(lambda x: [x[0], preprocess(x[1])])
id_summary_lTnoSW.take(5)
```

```
id_word = id_summary_lTnoSW.flatMap(lambda x: [((x[0], w),1) for w in x[1]])
id_word.collect()
```

```
id_word_freq = id_word.reduceByKey(lambda x,y: x+y)
id_word_freq.collect()
```

```
term_freqs = id_word_freq.map(lambda x: (x[0][1], (x[0][0], x[1])))
term_freqs.collect()
```

```
doc_freq = term_freqs.map(lambda x: (x[0], 1))
doc_freq = doc_freq.reduceByKey(lambda x,y: x+y)
doc_freq.collect()
```

```
import math
```

```
inv_doc_freq = doc_freq.map(lambda x: (x[0], 1+math.log(n/x[1])))
inv_doc_freq.collect()
```

```
tf_idf = term_freqs.join(inv_doc_freq)
tf_idf.collect()
```

```
tf_idf_prod = tf_idf.map(lambda x: (x[1][0][0], (x[0], x[1][0][1], x[1][1], x[1][0][1]*x[1][1])))
tf_idf_prod.collect()
```

```
prod_movies = tf_idf_prod.join(id_movies)
prod_movies.collect()
```

```
search = 'good cop bad cop'
q = search.split()
q_tf = sc.parallelize(q).map(lambda x:(x,1)).reduceByKey(lambda x,y: x+y)
q_tf.collect()
```

```
tf_idf_select = tf_idf.filter(lambda x: x[0] in q)
tf_idf_select.collect()
```

```
query_tfidf = q_tf.join(tf_idf_select)
# query_tfidf = tf_idf_select.join(broadcast(q_tf))
query_tfidf.collect()
```

```
query_tf_idf = query_tfidf.map(lambda x:(x[0],x[1][0]*x[1][1][1]))
# query_tf_idf = query_tf_idf.map(lambda x: ( (x[0], x[1])))
query_tf_idf = query_tf_idf.distinct()
query_tf_idf.collect()
```

```
query_doc_tf_idf = prod_movies.filter(lambda x: x[1][0][0] in q)
query_doc_tf_idf = query_doc_tf_idf.map(lambda x: (x[1][0][0], (x[1][1], x[1][0][3])))
query_doc_tf_idf.collect()
```

```
query_doc_tf_idf = query_tf_idf.join(query_doc_tf_idf)
# query_doc_tf_idf = query_doc_tf_idf.join(query_tf_idf)
query_doc_tf_idf.collect()
```

```
query_doc_tf_idf1 = query_doc_tf_idf.map(lambda x : (x[1][1][0], x[1][0], x[1][1][1]))
# query_doc_tf_idf = query_doc_tf_idf.map(lambda x: (x[0], (x[1]*x[2], x[2]*x[2], x[1]*x[1])))
query_doc_tf_idf1.collect()
```

```
query_doc_tf_idf2 = query_doc_tf_idf1.map(lambda x: (x[0], (x[1]*x[2], x[2]*x[2], x[1]*x[1])))
query_doc_tf_idf2.collect()
```

```
query_doc_tf_idf3 = query_doc_tf_idf2.reduceByKey(lambda x,y : ((x[0] + y[0], x[1] + y[1], x[2] + y[2])))
query_doc_tf_idf3.collect()
```

```
cos_sim = query_doc_tf_idf3.map(lambda x: (x[0], x[1][0]/(math.sqrt(x[1][1])*math.sqrt(x[1][2]))))
cos_sim.collect()
```

```
cos_sim2 = cos_sim.sortBy(keyfunc = lambda x: x[1], ascending = False).map(lambda x: x[0]).take(30)
for movie in cos_sim2:
    print(movie)
```

```
search_terms = sc.textFile("dbfs:/FileStore/shared_uploads/asi200000@utdallas.edu/Search_Terms.txt")
search_terms.collect()
```

```
search_terms = search_terms.map(lambda x: x.lower())
for search in search_terms.collect():
    print('\nSeach term:', search)
    q = search.split()
    if len(q)==1:
        # tf-idf scores
        result = prod_movies.filter(lambda x : q[0]==x[1][0][0]).sortBy(keyfunc = lambda x: x[1][0][3], ascending = False).map(lambda x: x[1][1]).take(10)
    else:
        # cosine similarity
        q_tf = sc.parallelize(q).map(lambda x:(x,1)).reduceByKey(lambda x,y: x+y)
        tf_idf_select = tf_idf.filter(lambda x: x[0] in q)
        query_tfidf = q_tf.join(tf_idf_select)
        query_tf_idf = query_tfidf.map(lambda x:(x[0],x[1][0]*x[1][1][1]))
        query_tf_idf = query_tf_idf.distinct()
        query_doc_tf_idf = prod_movies.filter(lambda x: x[1][0][0] in q)
        query_doc_tf_idf = query_doc_tf_idf.map(lambda x: (x[1][0][0], (x[1][1], x[1][0][3])))
        query_doc_tf_idf0 = query_tf_idf.join(query_doc_tf_idf)
        query_doc_tf_idf1 = query_doc_tf_idf0.map(lambda x : (x[1][1][0], x[1][0], x[1][1][1]))
        query_doc_tf_idf2 = query_doc_tf_idf1.map(lambda x: (x[0], (x[1]*x[2], x[2]*x[2], x[1]*x[1])))
        query_doc_tf_idf3 = query_doc_tf_idf2.reduceByKey(lambda x,y : ((x[0] + y[0], x[1] + y[1], x[2] + y[2])))
        cos_sim = query_doc_tf_idf3.map(lambda x: (x[0], x[1][0]/(math.sqrt(x[1][1])*math.sqrt(x[1][2]))))
        result = cos_sim.sortBy(keyfunc = lambda x: x[1], ascending = False).map(lambda x: x[0]).take(10)
    print('Movies:')
    for movie in result:
        print('\t', movie)
```
