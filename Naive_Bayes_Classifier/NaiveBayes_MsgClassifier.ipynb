```
data = sc.textFile("dbfs:/FileStore/shared_uploads/asi200000@utdallas.edu/SMSSpamCollection")
data.collect()
```

```
separated = data.map(lambda x: x.split('\t'))
separated.collect()
```

```
from nltk.tokenize import wordpunct_tokenize
```

```
prep0 = separated.map(lambda x: (x[0], wordpunct_tokenize(x[1].lower())))
prep0.collect()
```

```
import nltk
from nltk.corpus import stopwords
```

```
stopwrds = set(stopwords.words("english"))
```

```
prep1 = prep0.map(lambda x: (x[0], [y for y in x[1] if y not in stopwrds]))
prep1.collect()
```

```
from string import punctuation
```

```
prep2 = prep1.map(lambda x: (x[0], [y for y in x[1] if y not in punctuation]))
prep2.collect()
```

```
from nltk.stem.porter import PorterStemmer
```

```
stemmer = PorterStemmer()
```

```
prep3 = prep2.map(lambda x: (x[0], [stemmer.stem(y) for y in x[1]]))
prep3.collect()
```

```
trainingData, testData = prep3.randomSplit([0.75, 0.25])
```

```
num = trainingData.count()
num
```

```
labelCounts = trainingData.map(lambda x: (x[0], 1)).reduceByKey(lambda x, y: x + y)
labelCounts.collect()
```

```
import math
```

```
priors = labelCounts.map(lambda x: (x[0], math.log(x[1]/num)))
priors.collect()
```

```
priorMap = priors.collectAsMap()
print(priorMap)
```

```
vector_label = trainingData.flatMap(lambda x: [((x[0], y), 1) for y in x[1]]).reduceByKey(lambda x,y: x+y)
vector_label.collect()
```

```
class_words = vector_label.map(lambda x: (x[0][0], {x[0][1]: x[1]})).reduceByKey(lambda x, y: {**x, **y}).collectAsMap()
print(class_words)
```

```
prob_word_given_category = {}
for category, tf in class_words.items():
    class_counts = sum(tf.values())
    conditional_prob = {}
    for word, freq in tf.items():
        conditional_prob[word] = math.log(freq/class_counts)
    prob_word_given_category[category] = conditional_prob
print(prob_word_given_category)
```

```
def categorize(sample):
    cat_scores = {}
    for category, prior in priorMap.items():
        probs = prob_word_given_category[category]
        likelihood = prior
        for stem in sample:
            likelihood += probs.get(stem, 0)
        cat_scores[category] = likelihood
    return(min(cat_scores, key=cat_scores.get))
```

```
predictions = testData.map(lambda x: (x[0], categorize(x[1]), x[1]))
predictions.collect()
```

```
conf_matrix = predictions.map(lambda x: (x[0], x[1]))
conf_matrix.collect()
```

```
tn = conf_matrix.filter(lambda x: x[0]=='spam' and x[1]=='spam').count()
fn = conf_matrix.filter(lambda x: x[0]=='ham' and x[1]=='spam').count()
fp = conf_matrix.filter(lambda x: x[0]=='spam' and x[1]=='ham').count()
tp = conf_matrix.filter(lambda x: x[0]=='ham' and x[1]=='ham').count()
print('\t\t Real Ham \t Real Spam')
print('Pred Ham \t', tp, '\t\t', fp)
print('Pred Spam \t', fn, '\t\t', tn)
```

```
accuracy = 1.0*(tp+tn)/(tp+fp+fn+tn)
precision = 1.0*tp/(tp+fp)
recall = 1.0*tp/(tp+fn)
fmeasure = 2.0*precision*recall/(precision+recall)
print('Accuracy =', accuracy)
print('Precision =', precision)
print('Recall =', recall)
print('F-measure =', fmeasure)
```
